```python
def more(df):
    columns_radiant = [x for x in df.columns.tolist() if x.startswith('r1')  or
                       x.startswith('r2') or x.startswith('r3') or x.startswith('r4') or x.startswith('r5')]
    columns_dire = [x for x in df.columns.tolist() if x.startswith('d1')  or
                       x.startswith('d2') or x.startswith('d3') or x.startswith('d4') or x.startswith('d5')]
    size = len(columns_radiant)
    for a in range(size):
        df['more_' + str(a)] = df[columns_radiant[a]] - df[columns_dire[a]]
    return df
```
